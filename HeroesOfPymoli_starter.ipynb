## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins =[1,9.9,14.9,19.9,24.9,29.9,34.9,39.9,1000]
group_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data['Group Ages']=pd.cut(purchase_data['Age'],bins, labels=group_names,include_lowest=True)
group = purchase_data.groupby('Group Ages').count()
groupdrop = group.drop(['SN','Age','Gender','Item_ID','Item_Name','Price'],axis=1)
groupdrop.columns = ['Totals']
groupdrop['Percentage'] = groupdrop.Totals / groupdrop.Totals.sum()
groupdrop.Percentage = groupdrop['Percentage'].astype(float).map("{:.2%}".format)
groupdrop
```
